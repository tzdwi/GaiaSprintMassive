```python
import numpy as np, pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
blue_features = pd.read_csv('bright_blue_features.csv')
bright_sample = pd.read_csv('../data/bright_sample/bright_clean_w1w2_gt3.csv.gz')
```

```python
blue_sample = bright_sample[bright_sample['phot_bp_mean_mag'] - bright_sample['phot_rp_mean_mag'] < 1]
```

```python
timeout_names = np.setdiff1d(blue_sample['original_ext_source_id'].values,blue_features['Name'].values)
timeout_names
#first is just a dense lightcurve, second is too short.
```

```python
blue_good = blue_features.dropna()

blue = pd.merge(blue_sample,blue_good,left_on='original_ext_source_id',right_on='Name')
```

```python
#CMD of returned objects
plt.hexbin(blue['w1mpro']-blue['w2mpro'],blue['w1mpro'] + 5 * np.log10(blue['parallax']) - 10,mincnt=1,bins='log')

#plt.xlim(-0.21,1)
plt.ylim(1,-18)
plt.xlabel('W1 - W2')
plt.ylabel('M_W1')
plt.colorbar(label='log N')
```
