```python
import numpy as np, pandas as pd
from glob import glob
from matplotlib import pyplot as plt
from scipy.optimize import minimize
from WISE_tools import *
from astropy.stats import LombScargle

%matplotlib inline
```

```python
unique_names = parse_source_names('test_data/')
```

```python
#Try accessing and concatenating lightcurves
for name in unique_names[1:]:
    print(name)
    
    df = get_lightcurve(name, 'test_data/')
    
    plt.errorbar(df['mjd'], df['w1mpro'], yerr=df['w1sigmpro'],
                 marker='o', linestyle='none', alpha=0.25)
plt.gca().invert_yaxis()
```

```python
mjds = []
W1s = []
W1errs = []
for name in unique_names:
    df = get_lightcurve(name, 'test_data/')
    mjd,W1,W1err = df['mjd'].values, df['w1mpro'].values, df['w1sigmpro'].values
    mjds.append(mjd)
    W1s.append(W1)
    W1errs.append(W1err)
```

```python
W1_f, W1_p = LombScargle(mjd, W1, W1err).autopower()
```

```python
plt.plot(1/W1_f, W1_p)
plt.xscale('log')
```

```python
from astroML.time_series import ACF_EK
```

```python
for mjd,W1,W1err,name,c in zip(mjds,W1s,W1errs,unique_names,['C0','C2']):
    C_EK, C_EK_err, bins = ACF_EK(mjd, W1, W1err, bins=np.linspace(0, (max(mjd)-min(mjd))/2, 50))
    t_EK = 0.5 * (bins[1:] + bins[:-1])
    
    p = np.polyfit(t_EK[~np.isnan(C_EK)], C_EK[~np.isnan(C_EK)], deg=1, w=1/C_EK_err[~np.isnan(C_EK)])
    t_fit = np.linspace(np.min(t_EK), np.max(t_EK),100)
    fit = np.polyval(p, t_fit)
    AC_time = (np.exp(-1) - p[1])/p[0]
    print(AC_time)
    plt.axvline(x=AC_time, c=c, ls='--')
    plt.errorbar(t_EK, C_EK, C_EK_err, c=c, fmt = '.', lw=1, label=name)
    plt.plot(t_fit, fit, ls = '-.', c=c, lw=1, label='fit, '+name)
    plt.legend(loc=3)
plt.axhline(y=np.exp(-1))
```
